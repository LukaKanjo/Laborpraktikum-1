### Reale Spannungsquelle
#### Prinzip
Eine reale Spannungsquelle hat einen Spannugsabfall mit zunahme der Stromstärk/Belastung des Systems. Dieser Abfall lässt sich als Innenwiderstand interpretieren. daraus folgt das die Klemmenspannung $U_{KL}$ wie folgt ausgedrückt werden kann:
$$U_{KL}=U_0-IR_i$$

Die Klemmspannung nimmt dementsprechend linear ab.

#### Aufbau und Durchführung

Um den Innenwiderstand einer Baterie zu messen, wird ein variabler Widerstand in reihe Geschaltet. Daudurch kann die Stromstärke im Kreis reguliert werden. dabei werden die Klemmspannung und die Klemmstärke des Stromkreises gemessen.

#### Ergebnisse
Das Experiment zeigt ,dass ein Spannungsabfall in dieser Form vorliegt. Die Quellenspannung, ohne Lastwiderstand, beträgt $1.571\pm 0.080$. Der Innenwiderstand beträgt $0.322\pm 0.019$. Diese wurde auf 5% gerundet, da die relativen Unsicherheiten, der Messwerte, unter Fehlerfortpflanzung knapp 3% ergibt und es im Aufbau einen systematischen Fehler gibt, da wir die Innenwiderstände der Messgeräte ignorieren.
